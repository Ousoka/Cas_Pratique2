2. En se basant sur le tableau des coefficients, écrire un programme qui cacule la moyenne de chaque UE et détermine si l'étudiant l'a validée ou pas.

```python
P1= deliberation.iloc[:, 2]*(2.5/6)
P2= deliberation.iloc[:, 3]*(2.5/6)
TPP= deliberation.iloc[:, 4]*(1/6)
P=P1+P2+TPP
#print(P)

M1= deliberation.iloc[:, 5]*(2.5/5)
M2= deliberation.iloc[:, 6]*(2.5/5)
M=M1+M2
#print(M)

C1= deliberation.iloc[:, 7]*(2.5/6)
C2= deliberation.iloc[:, 8]*(2.5/6)
TPC= deliberation.iloc[:, 9]*(1/6)
C=C1+C2+TPC
#print(C)

I1= deliberation.iloc[:, 10]*(2.5/5)
I2= deliberation.iloc[:, 11]*(2.5/5)
I=I1+I2
#print(I)

B1= deliberation.iloc[:, 12]*(1.5/4)
B2= deliberation.iloc[:, 13]*(1.5/4)
B3= deliberation.iloc[:, 14]*(1/4)
B=B1+B2+B3
#print(B)

G1= deliberation.iloc[:, 15]*(2/4)
G2= deliberation.iloc[:, 16]*(2/4)
G=G1+G2
#print(G)
```

```python
DP=[]
dp=[]
for i in range(len(P)):
    if P[i]>=10:
        DP.append("V")
        dp.append(12)
    else:
        DP.append("NV")
        dp.append(0)
DM=[]
dm=[]
for i in range(len(P)):
    if M[i]>=10:
        DM.append("V")
        dm.append(10)
    else:
        DM.append("NV")
        dm.append(0)
DC=[]
dc=[]
for i in range(len(P)):
    if C[i]>=10:
        DC.append("V")
        dc.append(12)
    else:
        DC.append("NV")
        dc.append(0)
DI=[]
di=[]
for i in range(len(P)):
    if I[i]>=10:
        DI.append("V")
        di.append(10)
    else:
        DI.append("NV")
        di.append(0)
DB=[]
db=[]
for i in range(len(P)):
    if B[i]>=10:
        DB.append("V")
        db.append(8)
    else:
        DB.append("NV")
        db.append(0)
DG=[]
dg=[]
for i in range(len(P)):
    if G[i]>=10:
        DG.append("V")
        dg.append(8)
    else:
        DG.append("NV")
        dg.append(0)
Credits=[]
for i in range(len(dp)):
    k=dp[i]+dm[i]+dc[i]+di[i]+db[i]+dg[i]
    Credits.append(k)
Resultats=[]
for i in range(len(Credits)):
    if Credits[i]>=10 :
        Resultats.append("V")
    else:
        Resultats.append("NV")
```

3. Écrire un programme qui insère une colonne de validation (**V** : UE validée, **NV** : UE non validée) après chaque UE.

```python
deliberation1 = deliberation
deliberation1.insert(5, 'Moy_Physique', P, allow_duplicates = False)
deliberation1.insert(6, 'Res_Physique', DP, allow_duplicates = False)
deliberation1.insert(9, 'Moy_Maths', M, allow_duplicates = False)
deliberation1.insert(10, 'Res_Maths', DM, allow_duplicates = False)
deliberation1.insert(14, 'Moy_Chimie', C, allow_duplicates = False)
deliberation1.insert(15, 'Res_Chimie', DC, allow_duplicates = False)
deliberation1.insert(18, 'Moy_Informatique', I, allow_duplicates = False)
deliberation1.insert(19, 'Res_Informatique', DI, allow_duplicates = False)
deliberation1.insert(23, 'Moy_Base', B, allow_duplicates = False)
deliberation1.insert(24, 'Res_Base', DB, allow_duplicates = False)
deliberation1.insert(27, 'Moy_Graphique', G, allow_duplicates = False)
deliberation1.insert(28, 'Res_Graphique', DG, allow_duplicates = False)
deliberation1.insert(29, 'TotalCredits', Credits, allow_duplicates = False)
deliberation1.insert(30, 'Décision', Resultats, allow_duplicates = False)
deliberation1
```
