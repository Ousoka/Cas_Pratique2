2. En se basant sur le tableau des coefficients, écrire un programme qui cacule la moyenne de chaque UE et détermine si l'étudiant l'a validée ou pas.

```python
P1= deliberation.iloc[:, 2]*(2.5/6)
P2= deliberation.iloc[:, 3]*(2.5/6)
TPP= deliberation.iloc[:, 4]*(1/6)
P=P1+P2+TPP
#print(P)

M1= deliberation.iloc[:, 5]*(2.5/5)
M2= deliberation.iloc[:, 6]*(2.5/5)
M=M1+M2
#print(M)

C1= deliberation.iloc[:, 7]*(2.5/6)
C2= deliberation.iloc[:, 8]*(2.5/6)
TPC= deliberation.iloc[:, 9]*(1/6)
C=C1+C2+TPC
#print(C)

I1= deliberation.iloc[:, 10]*(2.5/5)
I2= deliberation.iloc[:, 11]*(2.5/5)
I=I1+I2
#print(I)

B1= deliberation.iloc[:, 12]*(1.5/4)
B2= deliberation.iloc[:, 13]*(1.5/4)
B3= deliberation.iloc[:, 14]*(1/4)
B=B1+B2+B3
#print(B)

G1= deliberation.iloc[:, 15]*(2/4)
G2= deliberation.iloc[:, 16]*(2/4)
G=G1+G2
#print(G)
```

```python
deliberation1 = deliberation.iloc[:, 0:2]
deliberation1.insert(2, 'Physique', P, allow_duplicates = False)
deliberation1.insert(3, 'Maths', M, allow_duplicates = False)
deliberation1.insert(4, 'Chimie', C, allow_duplicates = False)
deliberation1.insert(5, 'Informatique', I, allow_duplicates = False)
deliberation1.insert(6, 'Base', B, allow_duplicates = False)
deliberation1.insert(7, 'Graphique', G, allow_duplicates = False)
deliberation1
```

```python
DPk=[]
for i in range(len(P)):
    if P[i]>=10:
        DPk.append("Validé")
    else:
        DPk.append("Non Validé")
DMk=[]
for i in range(len(P)):
    if M[i]>=10:
        DMk.append("Validé")
    else:
        DMk.append("Non Validé")
DCk=[]
for i in range(len(P)):
    if C[i]>=10:
        DCk.append("Validé")
    else:
        DCk.append("Non Validé")
DIk=[]
for i in range(len(P)):
    if I[i]>=10:
        DIk.append("Validé")
    else:
        DIk.append("Non Validé")
DBk=[]
for i in range(len(P)):
    if B[i]>=10:
        DBk.append("Validé")
    else:
        DBk.append("Non Validé")
DGk=[]
for i in range(len(P)):
    if G[i]>=10:
        DGk.append("Validé")
    else:
        DGk.append("Non Validé")
```

```python
deliberation3 = deliberation.iloc[:, 0:2]
deliberation3.insert(2, 'Physique', DPk, allow_duplicates = False)
deliberation3.insert(3, 'Maths', DMk, allow_duplicates = False)
deliberation3.insert(4, 'Chimie', DCk, allow_duplicates = False)
deliberation3.insert(5, 'Informatique', DIk, allow_duplicates = False)
deliberation3.insert(6, 'Base', DBk, allow_duplicates = False)
deliberation3.insert(7, 'Graphique', DGk, allow_duplicates = False)
deliberation3
```
